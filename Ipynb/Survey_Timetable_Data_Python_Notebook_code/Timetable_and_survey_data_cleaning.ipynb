#### Second method

```python
def convert_timetable(timetable, room, df):
    for i in range(9):
        timetable.loc[len(timetable)] = [room, "02/11/2015", 
                                        df.iloc[i+1][0], 
                                        df.iloc[i+1][1],
                                        df.iloc[i+1][2]]
        timetable.loc[len(timetable)] = [room, "03/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][3],
                                        df.iloc[i+1][4]]
        timetable.loc[len(timetable)] = [room, "04/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][5],
                                        df.iloc[i+1][6]]
        timetable.loc[len(timetable)] = [room, "05/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][7],
                                        df.iloc[i+1][8]]
        timetable.loc[len(timetable)] = [room, "06/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][9],
                                        df.iloc[i+1][10]]
        timetable.loc[len(timetable)] = [room, "09/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][13],
                                        df.iloc[i+1][14]]
        timetable.loc[len(timetable)] = [room, "10/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][15],
                                        df.iloc[i+1][16]]
        timetable.loc[len(timetable)] = [room, "11/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][17],
                                        df.iloc[i+1][18]]
        timetable.loc[len(timetable)] = [room, "12/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][19],
                                        df.iloc[i+1][20]]
        timetable.loc[len(timetable)] = [room, "13/11/2015", 
                                        df.iloc[i+1][0],
                                        df.iloc[i+1][21],
                                        df.iloc[i+1][22]]
    return timetable
```

### Transforming the date and time to EPOCH format

```python
occupancy_df['event_time'] = None
occupancy_df['date2'] = None
occupancy_df['hour2'] = None
```

```python
for i in range(occupancy_df.shape[0]):
    if occupancy_df['hour'][i] == '9.00-10.00':
        occupancy_df.set_value(i,"hour2","9:00:00")
    elif occupancy_df['hour'][i] == '10.00-11.00':
        occupancy_df.set_value(i,"hour2","10:00:00")
    elif occupancy_df['hour'][i] == '11.00-12.00':
        occupancy_df.set_value(i,"hour2","11:00:00")
    elif occupancy_df['hour'][i] == '12.00-13.00':
        occupancy_df.set_value(i,"hour2","12:00:00")
    elif occupancy_df['hour'][i] == '13.00-14.00':
        occupancy_df.set_value(i,"hour2","13:00:00")
    elif occupancy_df['hour'][i] == '14.00-15.00':
        occupancy_df.set_value(i,"hour2","14:00:00")
    elif occupancy_df['hour'][i] == '15.00-16.00':
        occupancy_df.set_value(i,"hour2","15:00:00")
    elif occupancy_df['hour'][i] == '16.00-17.00':
        occupancy_df.set_value(i,"hour2","16:00:00")
    else:
        raise ValueError("invalid time")
        
    if occupancy_df['date'][i] == '2nd November 2015':
        occupancy_df.set_value(i,"date2","02/11/2015")
    elif occupancy_df['date'][i] == '3rd November 2015':
        occupancy_df.set_value(i,"date2","03/11/2015")
    elif occupancy_df['date'][i] == '4th November 2015':
        occupancy_df.set_value(i,"date2","04/11/2015")
    elif occupancy_df['date'][i] == '5th November 2015':
        occupancy_df.set_value(i,"date2","05/11/2015")
    elif occupancy_df['date'][i] == '6th November 2015':
        occupancy_df.set_value(i,"date2","06/11/2015")
    elif occupancy_df['date'][i] == '9th November 2015':
        occupancy_df.set_value(i,"date2","09/11/2015")
    elif occupancy_df['date'][i] == '10th November 2015':
        occupancy_df.set_value(i,"date2","10/11/2015")
    elif occupancy_df['date'][i] == '11th November 2015':
        occupancy_df.set_value(i,"date2","11/11/2015")
    elif occupancy_df['date'][i] == '12th November 2015':
        occupancy_df.set_value(i,"date2","12/11/2015")
    elif occupancy_df['date'][i] == '13th November 2015':
        occupancy_df.set_value(i,"date2","13/11/2015")
    else:
        raise ValueError("invalid date")
```

```python
occupancy_df['date3'] = None

for i in range(occupancy_df.shape[0]):
    occupancy_df.set_value(i,"date3", occupancy_df["date2"][i] + " " + occupancy_df["hour2"][i])
```

```python
import time

for i in range(occupancy_df.shape[0]):
    occupancy_df.set_value(i,"event_time", int(time.mktime(time.strptime(occupancy_df["date3"][i], 
                                                                         '%d/%m/%Y %H:%M:%S'))))
```
